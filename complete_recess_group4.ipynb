```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

dataR = pd.read_csv('processed_data.csv')
df = pd.DataFrame(dataR)
```

```python
df.drop(columns=['Time', 'Date', 'Date Time'], inplace=True)
```

```python
# Log transformation
log_transformed_df = df.copy()
for column in log_transformed_df.columns:
    if column != 'Class ' and log_transformed_df[column].dtype != 'object':
        log_transformed_df[column] = log_transformed_df[column].apply(lambda x: np.log1p(x) if x >= 0 else np.log1p(0.001))

# Normalization
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(log_transformed_df.drop(columns=['Class ']))
scaled_df = pd.DataFrame(scaled_features, columns=log_transformed_df.columns[:-1])
scaled_df['Class '] = log_transformed_df['Class ']
```

```python

# Split the data into training and testing sets
X = scaled_df.drop(columns=['Class '])
y = scaled_df['Class ']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Random Forest Model
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python


# Evaluation
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print(class_report)

# Plot Confusion Matrix
plt.figure(figsize=(10,7))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load your dataset
dataKV = pd.read_csv('processed_data.csv')

```

```python
df =pd.DataFrame(dataKV)
```

```python
#removing duplicate values while keeping the occurrence
dataKV = dataKV.drop_duplicates()
#To remove all duplicates include
```

```python

# Data preprocessing
# Convert categorical feature 'Class' to numerical
df['Class '] = df['Class '].map({'F': 0, 'M': 1})

df.drop(columns=['Date', 'Time', 'Date Time'], inplace=True)

# Separate features and target variable
X = df.drop(columns=['Class '])
y = df['Class ']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the models
log_reg = LogisticRegression(max_iter=1000)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
svm_model = SVC()
knn_model = KNeighborsClassifier()


```

```python
# Perform cross-validation and print results
models = [('Logistic Regression', log_reg), ('Random Forest', rf_model), ('SVM', svm_model), ('KNN', knn_model)]
for name, model in models:
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f'{name} Cross-Validation Accuracy: {scores.mean()}')

  
```

```python
# Train and evaluate each model on the test set
for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    print(f'\n{name} Test Accuracy: {accuracy}')
    print('Confusion Matrix:')
    print(conf_matrix)
    print('Classification Report:')
    print(class_report)
```
