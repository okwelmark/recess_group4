```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load your dataset
dataKV = pd.read_csv('clean_card_data.csv')

```

```python
df =pd.DataFrame(dataKV)
```

```python
# #checking for duplicate rows
# duplicates = dataKV.duplicated(keep=False)
# dataK[duplicates]
```

```python
#removing duplicate values while keeping the occurrence
dataKV = dataKV.drop_duplicates()
#To remove all duplicates include
```

```python
# df = df.to_csv('clean_card_data.csv', index=False)
```

```python

# Data preprocessing
# Convert categorical feature 'Class' to numerical
df['Class '] = df['Class '].map({'F': 0, 'M': 1})

df.drop(columns=['Date', 'Time', 'Date Time'], inplace=True)

# Separate features and target variable
X = df.drop(columns=['Class '])
y = df['Class ']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the models
log_reg = LogisticRegression(max_iter=1000)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
svm_model = SVC()
knn_model = KNeighborsClassifier()


```

```python
# Perform cross-validation and print results
models = [('Logistic Regression', log_reg), ('Random Forest', rf_model), ('SVM', svm_model), ('KNN', knn_model)]
for name, model in models:
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f'{name} Cross-Validation Accuracy: {scores.mean()}')

  
```

```python
# Train and evaluate each model on the test set
for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    print(f'\n{name} Test Accuracy: {accuracy}')
    print('Confusion Matrix:')
    print(conf_matrix)
    print('Classification Report:')
    print(class_report)
```
